```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('spotify_clean.xlsx')
print(df)
```

```python
important_cols = ['genre', 'explicit_content', 'language', 'album', 'artist']

for col in important_cols:
    counts = df[col].value_counts()
    counts.plot(kind='barh', color='skyblue',xlabel='counts',ylabel='columns')
    plt.title(f'Distribution of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()
```

```python
numerical_cols = ['duration', 'popularity', 'stream']
for col in numerical_cols:
    plt.hist(df[col], bins=20, color='darkblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
language_counts = df['language'].value_counts()
plt.pie(language_counts, labels=language_counts.index, startangle=90, colors=['#2D3250','#FF6B6B'],shadow=True)
plt.title('Explicit vs Non-Explicit Songs')
plt.legend()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('spotify_clean.xlsx')
print(df)
```

```python

genre_summary = df.groupby('genre')[['popularity', 'stream']].mean()

fig, axis = plt.subplots(nrows=2, ncols=1, figsize=(10,8))
fig.suptitle('Average Popularity and Streams by Genre', fontsize=16)
fig.patch.set_facecolor('#f0f0f0')
fig.subplots_adjust(hspace=0.5)

axis[0].plot(genre_summary['popularity'], color='purple', marker='o')
axis[0].set_xticks(range(len(genre_summary)))
axis[0].set_xticklabels(genre_summary.index, rotation=45)
axis[0].set_title('Average Popularity by Genre')

axis[1].plot(genre_summary['stream'], color='orange', marker='o')
axis[1].set_xticks(range(len(genre_summary)))
axis[1].set_xticklabels(genre_summary.index, rotation=45)
axis[1].set_title('Average Streams by Genre')

plt.show()
```

```python
explicit_counts = df['explicit_content'].value_counts()
plt.pie(explicit_counts, labels=explicit_counts.index, autopct='%1.1f%%', startangle=90, colors=['#2D3250','#FF6B6B'])
plt.title('Explicit vs Non-Explicit Songs')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['duration'], df['popularity'], c='skyblue', alpha=0.6)
plt.title('Popularity vs Duration')
plt.xlabel('Duration (seconds)')
plt.ylabel('Popularity')
plt.show()

# Popularity vs Streams
plt.figure(figsize=(8,5))
plt.scatter(df['stream'], df['popularity'], c='orange', alpha=0.6)
plt.title('Popularity vs Streams')
plt.xlabel('Streams')
plt.ylabel('Popularity')
plt.show()
```

```python

numerical_cols = ['duration', 'popularity', 'stream']
for i in range(len(numerical_cols)):
    for j in range(i+1, len(numerical_cols)):
        plt.figure(figsize=(5,4))
        plt.scatter(df[numerical_cols[i]], df[numerical_cols[j]], alpha=0.5, color='skyblue')
        plt.xlabel(numerical_cols[i])
        plt.ylabel(numerical_cols[j])
        plt.title(f'{numerical_cols[i]} vs {numerical_cols[j]}')
        plt.show()
```

```python
top_genres = df['genre'].value_counts().head(10)
plt.bar(top_genres.index, top_genres.values, color='skyblue')
plt.xticks(rotation=45)
plt.title('Top 10 Genres by Number of Songs')
plt.ylabel('Number of Songs')
plt.xlabel('Genre')
plt.show()
```
